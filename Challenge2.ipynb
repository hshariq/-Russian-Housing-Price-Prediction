```python
import pandas as pd
import numpy as np
```

```python
#read csv files in data
train = pd.read_csv('train.csv')
test= pd.read_csv('test.csv')
```

```python
#only show rows that have null values
train[train.isnull().any(axis=1)]
test[test.isnull().any(axis=1)]
```

```python
#convert data types where coluhmns have values yes and no to 1 and 0
train = train.replace({'yes': 1, 'no': 0})
test = test.replace({'yes': 1, 'no': 0})
```

```python
#show unique values in train for object columns
for col in train.select_dtypes(include=['object']).columns:
    print(col, ':', train[col].unique(), 'labels')
```

```python
# Import the necessary libraries
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
lbl = LabelEncoder()

# Fit the label encoder on the combined data of train and test datasets
lbl.fit(list(train['ecology'].values) + list(test['ecology'].values))

# Convert the 'ecology' column in the train dataset
train['ecology'] = lbl.transform(list(train['ecology'].values))

# Convert the 'ecology' column in the test dataset
test['ecology'] = lbl.transform(list(test['ecology'].values))
```

```python
train.drop(columns=['sub_area'], inplace=True)
test.drop(columns=['sub_area'], inplace=True)
```

```python
train = pd.get_dummies(train, columns=['sub_area'], drop_first=True)
test = pd.get_dummies(test, columns=['sub_area'], drop_first=True)
```

```python
#do label encoding for prodcut type as well
lbl.fit(list(train['product_type'].values) + list(test['product_type'].values))
train['product_type'] = lbl.transform(list(train['product_type'].values))
test['product_type'] = lbl.transform(list(test['product_type'].values))
```

```python
import numpy as np
import pandas as pd

# Assuming df is your DataFrame
numeric_cols = train.select_dtypes(include=[np.number]).columns.tolist()

# Remove binary columns from the list
numeric_cols = [col for col in numeric_cols if len(train[col].unique()) > 2]

# Apply log transformation to highly skewed columns
for col in numeric_cols:
    if train[col].skew() > 0.6:  # Adjust this threshold as needed
        train[col] = np.log1p(train[col])
```

```python
import numpy as np
import pandas as pd

# Assuming df is your DataFrame
numeric_cols = test.select_dtypes(include=[np.number]).columns.tolist()

# Remove binary columns from the list
numeric_cols = [col for col in numeric_cols if len(test[col].unique()) > 2]

# Apply log transformation to highly skewed columns
for col in numeric_cols:
    if test[col].skew() > 0.6:  # Adjust this threshold as needed
        test[col] = np.log1p(test[col])
```

```python
X = train.drop(columns=['price_doc'])
y = train[['price_doc']]
```
