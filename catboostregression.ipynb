```python
from sklearn.preprocessing import MinMaxScaler

# Create a scaler object
scaler = MinMaxScaler()

# Fit and transform the numeric columns
X_scaled = scaler.fit_transform(X[numerical_features])

# Create a new DataFrame with the scaled numeric columns
X_scaled_df = pd.DataFrame(X_scaled, columns=numerical_features)

# Concatenate the scaled numeric columns with the categorical columns
X_catboost = pd.concat([X_scaled_df, train[categorical_features]], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_catboost, y, test_size=0.2, random_state=42)
```

```python
# Install CatBoost using: pip install catboost
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from catboost import CatBoostRegressor, Pool
import numpy as np


# Create a CatBoostRegressor
catboost_model = CatBoostRegressor(iterations=1500,  # Number of boosting iterations
                                   learning_rate=0.01,  # Step size shrinkage used in update to prevent overfitting
                                   depth=8,  # Depth of the trees
                                   loss_function='RMSE',  # Loss function for regression
                                   cat_features=categorical_features,
                                   random_seed=2)  # Random seed for reproducibility

# Create a CatBoost Pool for training and testing data
train_pool = Pool(X_train, y_train, cat_features=categorical_features)
test_pool = Pool(X_test, y_test, cat_features=categorical_features)

# Fit the model on the training data
# Fit the model on the training data
catboost_model.fit(train_pool, eval_set=(test_pool), verbose=2)

# catboost_model.fit(X_train, y_train, eval_set=(X_test, y_test), verbose=2)

# Make predictions on the test set
predictions = catboost_model.predict(test_pool)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
print("Root Mean squared error: %.2f" % rmse)

# You can also access feature importances
# feature_importances = catboost_model.feature_importances_
# print('Feature Importances:', feature_importances)
# bestTest = 12704833.59

```
